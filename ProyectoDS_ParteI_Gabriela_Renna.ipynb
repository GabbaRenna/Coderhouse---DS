```python
# Evaluamos la hipótesis "Hay mayor satisfacción laboral en empleos híbridos y remotos y menor satisfacción en presenciale" con gráfico de barras

# Corroboramos hipótesis siendo la modalidad 100% remoto la mejor calificada

promedio_por_modalidad = Encuesta_salarial.groupby('Modalidad')['Calificación del lugar de trabajo 1-10'].mean()
fig, ax = plt.subplots(figsize=(8,6))
ax.bar(promedio_por_modalidad.index, promedio_por_modalidad)
plt.xlabel('Modalidad')
plt.ylabel('Promedio de calificación')
plt.title('Relación entre modalidad de trabajo y la calificación del lugar de trabajo')
```
